# Cvičení 1 - Půlmaraton (alternativní řešení přes funkci pivot())

```python
import pandas as pd

data = pd.read_csv("half_marathon.csv")
data.head()
```

```python
data_pivot = data.pivot(values="Cas", columns=["Rok zavodu"], index=["Jmeno", "Rocnik"])
data_pivot["difference"] = data_pivot[2020] - data_pivot[2019]
data_pivot["difference"].mean()

data_pivot
```
