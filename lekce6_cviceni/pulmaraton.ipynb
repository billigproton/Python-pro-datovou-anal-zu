# Cvičení 1 - Půlmaraton

```python
import pandas as pd

data = pd.read_csv("half_marathon.csv")
data.head()
```

```python
data["Cas"] = pd.to_timedelta(data["Cas"])
data["Cas 2019"] = data["Cas"].shift(1)
data = data[data["Rok zavodu"] == 2020]
data["Zmena"] = data["Cas"] - data["Cas 2019"]
data
```

```python
data["Zmena"].mean()
```

```python
bins = [-float("inf"), 0, float("inf")]
labels = ["lepší než 2019", "horší než 2019"]
zavod['Porovnání 2019 a 2020'] = pd.cut(zavod["2020-2019"], bins=bins, labels=labels)

zavod
```
