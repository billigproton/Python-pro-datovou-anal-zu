# Alias sloupců v pandas (dočasný název pro sloupec) - obdoba AS v SQL

```python
import pandas as pd

jmena = pd.read_csv(r'C:\Users\Lenovo\OneDrive\Dokumenty\CZECHITAS-LAPTOP-BTGJD9J5\Python pro datovou analýzu\jmena.csv')

jmena
```

Sloupcům podobně jako v SQL alias dát jako takový nemohu, můžu si však v rámci kódu sloupec přejmenovávat. Musím ale brát ohled na to, že budu od té chvíle přejmenování pracovat s dataframem, kde už je to přejmenované - nutnost znova definovat názvy sloupců

Definuji si alias 'name' na sloupec 'jméno'

```python
jmena.columns = ["name", "cetnost", "vek", "pohlavi", "puvod"]

jmena
```

Pokud bych chtěl pracovat s původním názvem 'jméno', musím ho opět přejmenovat

```python
jmena.columns = ["jméno", "cetnost", "vek", "pohlavi", "puvod"]

jmena
```

Takže: přejmenovávat sloupce v pandas obecně jde, alias jako v SQL ne, ale můžu to obejít přes opakované přejmenovávání :)
