# Cvičení 1 - půlmaraton, řešení problému s nejasným výpočtem rozdílu

```python
import pandas as pd
```

Trochu jsem poupravil kód, ve výsledku si získáme sloupec se vteřinami. Nějaké máme kladné, nějaké záporné. To nám samo o sobě nic neřekne.

```python
import datetime as dt
zavod = pd.read_csv("lekce6_cviceni\half_marathon.csv")
zavod["date"] = pd.to_timedelta(zavod["Cas"])
zavod["vysledek_2019"] = zavod["date"].shift(1) #posun
zavod = zavod[zavod["Rok zavodu"]==2020]
zavod["2020-2019"] = (zavod["date"] - zavod["vysledek_2019"]).dt.total_seconds()
zavod
```

Problém můžeme vyřešit cestou vytvoření dvou kategorií, kdy nás zajímá, zda byl výsledek lepší než v roce 2019 nebo horší než v roce 2019, pomocí bins. Podle tohohle nového sloupce pak můžeme snadno vyfiltrovat ty, kteří se zlepšili/zhoršili :)

```python
bins = [-float("inf"), 0, float("inf")]
labels = ["lepší než 2019", "horší než 2019"]
zavod['Porovnání 2019 a 2020'] = pd.cut(zavod["2020-2019"], bins=bins, labels=labels)

zavod
```
